```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib notebook
```

```python
diabetic_data_pd =  pd.read_csv('/Users/tessa/Downloads/diabetic_data.csv')
#print(diabetic_data_pd)

diabetic_data_np = diabetic_data_pd.to_numpy()
```

```python
#Seaborn displot showing diabetic diagnosis distribution among age groups.

#Creating histogram, adjusted size to allow x-axis to not overlap age ranges.
diabetic_distribution = sns.displot(
   diabetic_data_pd['age'],color='r',height=5,aspect=1.4).set(title="Diabetic Age Distribution")   

#Adjusting figure to make room for title
diabetic_distribution.fig.subplots_adjust(top=.95) 

plt.show()
```

```python
#Using Seaborn displot to create a bivariate plot showing the relationship between age and the stay in hospitals.

#Creating bivariate plot
age_hospital_stay = sns.displot(
    diabetic_data_pd,x='time_in_hospital',y='age',color='plum',height=5,aspect=1.4).set(
    title="Age and time spent in hospital comparison due to Diabetes")   

#Inverting y-axis because it was listing the higher age numbers at the point where it meets the x-axis
plt.gca().invert_yaxis()

#Adjusting figure to make room for title
age_hospital_stay.fig.subplots_adjust(top=.95)

plt.show()
```

```python
#Pie chart showing admission types broken down into percentages

#Finding count of each admission type
admissiontype_diabetes = diabetic_data_np[:,6].copy()
unique,counts = np.unique(admissiontype_diabetes,return_counts=True)
admissiontype_counts = np.asarray((unique,counts))
#print(admissiontype_counts)

#Creating title and array of counts of admission types
title = plt.title("Admission Types")
admissiontype_pie = np.array([53990,18480,18869,10,4785,5291,21,320])

#Creating pie chart with labels
plt.gca().axis("equal")
explode = (0,0,0,0.9,0,0,0.9,0.1)
shorthand_labels = ['1','2','3','4','5','6','7','8']
admissiontype_labels = ['1:Emergency','2:Urgent','3:Elective','4:Newborn',
                        '5:Not Available','6:NULL','7:Trauma Center','8:Not Mapped']


plt.pie(admissiontype_pie, labels=shorthand_labels,explode=explode,autopct='%1.1F%%')
plt.legend(labels=admissiontype_labels,loc='best',bbox_to_anchor=(0.2,1))
plt.show()
```
